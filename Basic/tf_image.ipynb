```
import tensorflow as tf
```

```
x = tf.placeholder(tf.float32,shape=[None,32,32,3])
y_true = tf.placeholder(tf.float32,shape=[None,10])
```

```
hold_prob = tf.placeholder(tf.float32)
```

```
## creating 1st convolutional layer 
```

```
init_random_dist = tf.truncated_normal([4,4,3,32], stddev=0.1)
W=tf.Variable(init_random_dist)

init_bias_vals = tf.constant(0.1, shape=[[4,4,3,32][3]])
b=tf.Variable(init_bias_vals)

convo_1=tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')

convo_1=tf.nn.relu(convo_1 + b)

convo_1_pooling=tf.nn.max_pool(convo_1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
```

```
## creating 2st convolutional layer 
```

```
init_random_dist = tf.truncated_normal([4,4,32,64], stddev=0.1)
W=tf.Variable(init_random_dist)

init_bias_vals = tf.constant(0.1, shape=[[4,4,32,64][3]])
b=tf.Variable(init_bias_vals)

convo_2=tf.nn.conv2d(convo_1_pooling, W, strides=[1, 1, 1, 1], padding='SAME')

convo_2=tf.nn.relu(convo_2 + b)

convo_2_pooling=tf.nn.max_pool(convo_2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
```

```
convo_2_flat = tf.reshape(convo_2_pooling,[-1,8*8*64])
```

```
input_size = int(convo_2_flat.get_shape()[1])

init_random_dist = tf.truncated_normal([input_size,1024], stddev=0.1)
W=tf.Variable(init_random_dist)

init_bias_vals = tf.constant(0.1, shape=[1024])
b=tf.Variable(init_bias_vals)

full_layer_one=tf.nn.relu(tf.matmul(convo_2_flat, W) + b)
```

```
full_one_dropout = tf.nn.dropout(full_layer_one,keep_prob=hold_prob)
```

```
input_size = int(full_one_dropout.get_shape()[1])

init_random_dist = tf.truncated_normal([input_size,10], stddev=0.1)
W=tf.Variable(init_random_dist)

init_bias_vals = tf.constant(0.1, shape=[10])
b=tf.Variable(init_bias_vals)
y_pred=tf.matmul(full_one_dropout, W) + b
```

```
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_true,logits=y_pred))
```

```
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
train = optimizer.minimize(cross_entropy)
```
