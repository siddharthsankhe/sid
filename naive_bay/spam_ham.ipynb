```python
import numpy as np
import pandas as pd
```

```python
docs = pd.read_table('SMSSpamCollection', header=None, names=['Class', 'sms'])
```

```python
docs.head(2)
```

```python
docs['Class'].value_counts()
```

```python
docs['label'] = docs['Class'].map({'ham':0, 'spam':1})
```

```python
docs.head(2)
```

```python
X=docs['sms']
```

```python
y=docs['label']
```

```python
print(X.shape)
print(y.shape)
```

```python

from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
##assign each words in ascending order
vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
vect.get_feature_names()## sorted words 
```

```python
len(vect.get_feature_names())
```

```python
vect.get_feature_names().index('000')
```

```python
X_train_transformed = vect.transform(X_train)
```

```python
X_train_transformed.toarray()
```

```python
X_test_tranformed =vect.transform(X_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb = BernoulliNB()
bnb.fit(X_train_transformed,y_train)
y_pred_class = bnb.predict(X_test_tranformed)
y_pred_proba =bnb.predict_proba(X_test_tranformed)
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
print("PRECISION SCORE :",metrics.precision_score(y_test, y_pred_class))
print("RECALL SCORE :", metrics.recall_score(y_test, y_pred_class))
print("F1 SCORE :",metrics.f1_score(y_test, y_pred_class))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X_train_transformed.todense()
```

```python
gauss = GaussianNB()

gauss.fit(X_train_transformed.todense(),y_train)
```

```python
y_pred_class = gauss.predict(X_test_tranformed.todense())
y_pred_proba =gauss.predict_proba(X_test_tranformed.todense())
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
print("PRECISION SCORE :",metrics.precision_score(y_test, y_pred_class))
print("RECALL SCORE :", metrics.recall_score(y_test, y_pred_class))
print("F1 SCORE :",metrics.f1_score(y_test, y_pred_class))
```
